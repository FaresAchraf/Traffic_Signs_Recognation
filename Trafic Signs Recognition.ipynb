```python

import os
import skimage
def load_data(data_directory):
    directories = [d for d in os.listdir(data_directory) 
                   if os.path.isdir(os.path.join(data_directory, d))]
    labels = []
    images = []
    for d in directories:
        label_directory = os.path.join(data_directory, d)
        file_names = [os.path.join(label_directory, f) 
                      for f in os.listdir(label_directory) 
                      if f.endswith(".ppm")]
        for f in file_names:
            images.append(skimage.data.load(f))
            labels.append(int(d))
    return images, labels

ROOT_PATH = "C:\\Users\Achraf\Downloads"
train_data_directory = os.path.join(ROOT_PATH, "deep\Training")
test_data_directory = os.path.join(ROOT_PATH, "deep\Testing")

images, labels = load_data(train_data_directory)
```

```python
def evaluation(Z):
    plt.style.use("ggplot")

    plt.figure(figsize=(20,8))
    plt.plot(Z.history["loss"], label="train_loss")
    plt.plot(Z.history["val_loss"], label="val_loss")
    plt.plot(Z.history['accuracy'], label="train_acc")
    plt.plot(Z.history['val_accuracy'], label="val_acc")
    plt.title("Training Loss and Accuracy on Dataset")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss/Accuracy")
    plt.legend(loc="lower left")
```

# improving accuracy using CNN 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
import numpy as np

images=np.array(images).reshape(-1,40,40,1)
#labels=np.array(labels)

model_cnn = Sequential()

model_cnn.add(Conv2D(32, (3,3), input_shape=images.shape[1:]))
model_cnn.add(Activation('relu'))
model_cnn.add(MaxPooling2D(pool_size=(2, 2)))

model_cnn.add(Conv2D(64, (1,1)))
model_cnn.add(Activation('relu'))
model_cnn.add(MaxPooling2D(pool_size=(1, 1)))



model_cnn.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors

model_cnn.add(Dense(550))
model_cnn.add(Activation('relu'))

model_cnn.add(Dense(250))
model_cnn.add(Activation('relu'))

model_cnn.add(Dense(62))
model_cnn.add(Activation('sigmoid'))

model_cnn.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
b=model_cnn.fit(images,np.array(labels),batch_size=32, epochs=4,validation_split=0.05)
```

```python
def accuracy(model_cnn):
    predict=model_cnn.predict(images_test.reshape(2520,40,40,1))
    x=0
    for i in range(0,2520):
        if np.argmax(predict[i]) == labels_test[i]:
            x=x+1

    print("accuracy=",x/2520)

    
accuracy(model_cnn)
```

# Transfer Learning ‚Äî Pre-trained CNN (vgg16)as a Feature Extractor Performance

```python
input_shape = (150, 150, 3)
from keras.applications import vgg16
from keras.models import Model
import keras

vgg = vgg16.VGG16(include_top=False, weights='imagenet', 
                                     input_shape=input_shape)

output = vgg.layers[-1].output
output = keras.layers.Flatten()(output)
vgg_model = Model(vgg.input, output)

vgg_model.trainable = False
for layer in vgg_model.layers:
    layer.trainable = False
    
import pandas as pd
pd.set_option('max_colwidth', -1)
layers = [(layer, layer.name, layer.trainable) for layer in vgg_model.layers]
pd.DataFrame(layers, columns=['Layer Type', 'Layer Name', 'Layer Trainable'])    
```

```python
images3, labels3 = load_data(train_data_directory)
images3 = [transform.resize(image, (150, 150)) for image in images3]
images_test3, labels_test3 = load_data(test_data_directory)
images_test3 = [transform.resize(image, (150, 150)) for image in images_test3]

def get_bottleneck_features(model3, input_imgs):
    features = model3.predict(input_imgs, verbose=0)
    return features
    
train_features_vgg = get_bottleneck_features(vgg_model, np.array(images3))
validation_features_vgg = get_bottleneck_features(vgg_model, np.array(images_test3))

print('Train Bottleneck Features:', train_features_vgg.shape, 
      '\tValidation Bottleneck Features:', validation_features_vgg.shape)
```

```python
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, InputLayer
from keras.models import Sequential
from keras import optimizers

input_shape = vgg_model.output_shape[1]


model3 = Sequential()
model3.add(InputLayer(input_shape=(input_shape,)))
model3.add(Dense(300, activation='relu', input_dim=input_shape))
model3.add(Dense(200, activation='relu'))
model3.add(Dense(128, activation='relu'))
model3.add(Dense(62, activation=tf.nn.softmax))

model3.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model3.summary()
```

```python
history_1 = model3.fit(train_features_vgg,np.array(labels),validation_data=(validation_features_vgg, np.array(labels_test)),epochs=6)
```

```python
print("accuracy = ",model3.evaluate(validation_features_vgg, np.array(labels_test))[1])
```
